````python
def get_dataset(size):

    df = pd.DataFrame()
    df['size'] = np.random.choice(['big','medium','small'], size)
    df['age'] = np.random.randint(1, 50, size)
    df['color'] = np.random.choice(['red','blue','yellow','green'], size)
    df['win'] = np.random.choice(['yes','no'], size)
    dates = pd.date_range('2020-01-01', '2022-12-31')
    df['date'] = np.random.choice(dates, size)
    df['probability'] = np.random.uniform(0, 1, size)
    return df

def set_dtypes(df):
    df['size'] = df['size'].astype('category')
    df['color'] = df['color'].astype('category')
    df['age'] = df['age'].astype('int16')
    df['win'] = df['win'].map({'yes':True, 'no': False})
    df['probability'] = df['probability'].astype('float32')
    return df
````

````python
print('Reading and writing CSV')
df = get_dataset(5_000_000)
df = set_dtypes(df)
%time df.to_csv('test.csv')
%time df_csv = pd.read_csv('test.csv')
````

````python
print('Reading and writing Pickle')
df = get_dataset(5_000_000)
df = set_dtypes(df)
%time df.to_pickle('test.pickle')
%time df_csv = pd.read_pickle('test.pickle')
````

````python
print('Reading and writing Parquet')
df = get_dataset(5_000_000)
df = set_dtypes(df)
%time df.to_parquet('/Users/kasia/Desktop/programowanie/big-data/test.parquet')
%time df_csv = pd.read_parquet('/Users/kasia/Desktop/programowanie/big-data/test.parquet')
````

````python
print('Reading and writing Feather')
df = get_dataset(5_000_000)
df = set_dtypes(df)
%time df.to_feather('/Users/kasia/Desktop/programowanie/big-data/test.feather')
%time df_csv = pd.read_feather('/Users/kasia/Desktop/programowanie/big-data/test.feather')
````
